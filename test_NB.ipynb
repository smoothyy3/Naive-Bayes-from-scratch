```python
import numpy as np
from naive_bayes import Naive_Bayes
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_iris
from math_functions import *
import matplotlib.pyplot as plt
```

# Compare sklearn.NB with NB from scratch

```python
iris = load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### train NB from scratch

```python
nb_scratch = Naive_Bayes(laplace_smoothing=True)

nb_scratch.fit(X_train, y_train)
```

### train sklearn.NB and print results

```python
nb_sklearn = GaussianNB()
nb_sklearn.fit(X_train, y_train)

y_pred_scratch = nb_scratch.predict(X_test)
y_pred_sklearn = nb_sklearn.predict(X_test)

from sklearn.metrics import accuracy_score

print("Accuracy of Naive Bayes from scratch:", accuracy_score(y_test, list(y_pred_scratch.values())))
print("Accuracy of sklearn Naive Bayes:", accuracy_score(y_test, y_pred_sklearn))
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(range(len(y_test)), y_test, c=y_pred_sklearn, cmap='coolwarm')
plt.title("sklearn Naive Bayes")

plt.subplot(1, 2, 2)
plt.scatter(range(len(y_test)), y_test, c=list(y_pred_scratch.values()), cmap='coolwarm')
plt.title("Scratch Naive Bayes")

plt.show()
```
